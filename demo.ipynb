````python
import requests
import time
````

````python
spark = SparkSession.builder.appName("BikeWeatherPipeline").getOrCreate()
````

````python
bike_data = requests.get("https://api.citybik.es/v2/networks/citi-bike-nyc").json()
````

````python
from pprint import pprint

stations = bike_data['network']['stations']

pprint(stations[0], sort_dicts=False)
````

````python
station_data = []
for station in stations:
# station = stations[0]
    stations_id = station['id']
    station_name = station['name']
    latitude = station['latitude']
    longitude = station['longitude']
    address = station_name
    has_ebikes = station['extra']['has_ebikes']
    slots = station['extra']['slots']

    data = {
        'station_id': stations_id,
        'station_name': station_name,
        'latitude': latitude,
        'longitude': longitude,
        'address': address,
        'has_ebikes': has_ebikes,
        'slots': slots
    }

    station_data.append(data)


````

````python
WEATHER_API = "https://api.open-meteo.com/v1/forecast?latitude=40.7143&longitude=-74.006&hourly=temperature_2m,precipitation,wind_speed_10m,cloudcover,relativehumidity_2m"

waether_data = requests.get(WEATHER_API).json()
````

````python
hourly = waether_data['hourly']
hourly
````

````python
for a in zip(
            hourly["time"],
            hourly["temperature_2m"],
            hourly["precipitation"],
            hourly["wind_speed_10m"],
            hourly["cloudcover"],
            hourly["relativehumidity_2m"]
        ):
    print(a)
````
